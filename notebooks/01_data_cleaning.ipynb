```python
import pandas as pd
DATA_PATH = "../data/raw/hospital_los.csv"

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df.info #info on types and missing values
```

```python
df.describe(include="all")
```

```python
# Missing values. Which columns have missing data and how much missing data there is. 
df.isna().sum()
```

```python
# Correcting columns to correct data types
df['vdate'] = pd.to_datetime(df['vdate'])
```

```python
#Convert rcount to numeric so that the ML model can work properly.
df['rcount'] = pd.to_numeric(df['rcount'], errors='coerce').astype('Int64')
```

```python
#Inspecting categorical columns
#Gender
df['gender'].value_counts()
```

```python
#Secondary diagnosis
df['secondarydiagnosisnonicd9'].value_counts().head()
```

```python
#Discharged
df['discharged'].value_counts()
```

```python
#Faculty 
df['facid'].value_counts().head()
```

```python
# 1. Drop discharge date to avoid leakage
df = df.drop('discharged', axis=1)

#Enconde gender to binary
df['gender'] = df['gender'].map({'F': 0, 'M': 1})

#One-hot encode facility ID
df = pd.get_dummies(df, columns=['facid'], prefix='fac', drop_first=True)

#Extract useful date features
df['month'] = df['vdate'].dt.month
df['dayofweek'] = df['vdate'].dt.dayofweek

```

```python
df = df.drop(columns=['vdate'], errors='ignore')
```

```python
df.info()
```

```python
# Inspection of NaN values in rcount
df['rcount'].isna().sum()
```

```python
#Replace the missing values with 0 and convert the column to a normal int type
df['rcount']= df['rcount'].fillna(0).astype('int64')
```

```python
#Check again NaN value count in rcount
df['rcount'].isna().sum()
```

```python
# Now we have our cleaned df and can save it
df.to_csv('../data/processed/cleaned_hospital_los.csv', index=False)
```
